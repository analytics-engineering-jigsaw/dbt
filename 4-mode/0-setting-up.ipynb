# Setting Up 

### Introduction

In this lesson, we'll connect our snowflake database to our Mode BI tool.  Let's get started.

### Connecting our Database

We'll begin by selecting a database to connect.  Let's choose the snowflake database.

<img src="./choose-snowflake.png" width="60%">

Then we can enter the credentials relevant to our snowflake account.

> Notice that for account name, we copy the beginning part of our url address to snowflake -- not including the `https://`.

<img src="./enter-creds.png" width="60%">

Now that we have this, we'll be taken to a page that shows both the schemas from our database, and a SQL console.  

We can see the related schemas on the right.

<img src="./snowflake-schemas.png" width="80%">

And then we ran our first SQL query.

```sql
select city, count(*) as amount from dbt_jigsawlabsstudent.companies_dim group by city order by amount desc
```

As we can see, our SQL query selects from our `companies_dim` table, calculating each city, and the number of companies in that city.

We can use these results to create a new chart.  To do so, we click on the green tab to the right.

> <img src="./chart-tab.png" width="50%">

And upon clicking on this tab, we are taken to the new chart screen.

> <img src="./adding-axes.png" width="80%">

Notice that under the fields, we see the two columns resulting from our query.

We have both a *dimensions* up top, and *measures* below.  The dimensions are for any of the textual fields, and the measures are for the any numerical data.

### Building our bar chart

For the bar chart, we would like to have the amount along the x-axis and the amount along the y-axis.

So we can drag and drop those two fields to the x and y axis slots.

Upon doing so, we'll see the following:

![sort-desc.png](sort-desc.png)

Then, click on sort descending, and we'll get to the following:

<img src="./sorted-results.png" width="70%">

### Letting mode do the work

Now we produced the graph above by aggregating the results in our SQL query.

```sql
select city, count(*) as amount from dbt_jigsawlabsstudent.companies_dim group by city order by amount desc
```

But we can also let mode perform the group bys.

<img src="./customer-dim.png" width="60%">

And then when we move to the chart, we can select the name dimension as our y axis, and then choose `count`.

<img src="./count-by-city.png" width="100%">

> If we select the green drop down, we'll see there are options for other aggregations like count, and count distinct.

### Summary

In this lesson, we saw how we can begin using using mode to display some of our data.  As we saw, we use mode by first connecting to our database.  And then, we saw that we can use the result of our SQL query to display our data.

Finally, we saw that mode can perform aggregations for us -- grouping by the specified column, and performing the chosen aggregation.
